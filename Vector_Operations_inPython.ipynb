### Addition of the vector

```python
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])

sum=vector_1+vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,5,1,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

### Substraction of Vector

```python
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])

difference=vector_1-vector_2
print(difference)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,-1,5,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

### Multiplying the Vector

```python
vector_1=np.asarray([0,0,2,3])
vector_2=2*vector_1

print(vector_2)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,4,6,scale_units='xy',angles='xy',scale=1,color='y')

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_1=np.asarray([0,0,2,3])
vector_2=-0.5*vector_1

print(vector_2)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,-1,-1.5,scale_units='xy',angles='xy',scale=1,color='y')

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

## **Dot Product of Vector**
**a.b=|a||b| cos**

```python
a = np.array([2, 3])
b = np.array([4, 4])

a_dot_b = np.dot(a,b)
```

```python
print(a_dot_b)
```

```python
a = np.array([40, 20, 32])
b = np.array([53, 24, 68])

a_dot_b = np.dot(a,b)
print(a_dot_b)
```

## **Cross Product of Vector**
**A  B =|A| |B| sin **

```python
a = np.array([2, 3])
b = np.array([4, 4])

a_x_b = np.cross(a,b)
print(a_x_b)
```

```python
a = np.array([40, 20, 32])
b = np.array([53, 24, 68])
a_x_b = np.cross(a,b)
print(a_x_b)
```

## **Projection of Vector**
**Projection of a vector on b vector**

image.png

```python
a = np.array([40, 20, 32])
b = np.array([53, 24, 68])

#magnitute of b vector
magnitude_b= np.sqrt(np.sum(b**2))

#projection of a vector on b
proj_a_on_b=(np.dot(a,b)/magnitude_b**2)*b
print('Projection of a vector on b vector : ', proj_a_on_b)
```
